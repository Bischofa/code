```python
#df = pd.read_excel('{}'.format('mse_ms1964.csv'))
#mse = df.loc[idx, 'mse']
mse = ['mse13161', 'mse13169']
master_df = pd.DataFrame()

for exam in mse:
        print(exam)
        if exam.startswith("mse"):
            dicom = glob("/data/henry6/antje/transition/MPRAGE_diff_positions_ms1964/{0}/E*/*".format(exam))
            #print(dicom)
            df = pd.DataFrame()
            for idx,DCM in enumerate(dicom):
                first_dicom = glob(DCM+"/E*.DCM")[0]
                print(first_dicom)
                df.loc[idx, 'MSEID'] = exam
                df.loc[idx, 'Series'] = DCM.split('/')[-1]
                df.loc[idx,"DCM Name"] = first_dicom
                
                cmd = ["dicom_hdr", "-sexinfo", first_dicom]
                proc = Popen(cmd, stdout=PIPE)
                output =[l.decode("latin-1").split() for l in proc.stdout.readlines()[:]]
                for line in output:
#                     print(line)
                    
                    if not line:
                        continue
                        
                    
                    newline = line[0]
                    
                    if "lScanRegionPosTra" in newline:
                        ScannerBedPosition = line[2]
                        print("Scanner Bed Position", ScannerBedPosition)
                        df.loc[idx, "Scanner Bed Position"] = ScannerBedPosition
                            
       
            master_df = master_df.append(df)
master_df = master_df.reset_index().drop(['index'], axis=1)
                  #"{0}".format(out))
```

```python
master_df.to_csv('TablePosition.csv')
```
