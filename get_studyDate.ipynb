```python
def csv_from_excel(excel_file):
    import xlrd
    import csv
    wb = xlrd.open_workbook(excel_file,encoding_override="cp1252")
    sh = wb.sheet_by_name('advanced search results')
    outfile = excel_file.replace(".xls",".csv")
    your_csv_file = open(outfile, 'wb')
    wr = csv.writer(your_csv_file, quoting=csv.QUOTE_ALL)

    for rownum in xrange(sh.nrows):
        try:
            wr.writerow(sh.row_values(rownum))
        except UnicodeEncodeError:
            print "Can't write row!!", rownum

    your_csv_file.close()
    return outfile
```

```python
all_dates_csv = csv_from_excel("/home/abischof/Downloads/all_study_dates.xls")
```
