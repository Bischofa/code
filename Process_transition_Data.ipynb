```python
import os
import pandas as pd
from subprocess import check_call
```

```python
def _get_output(mse):
# use this to return which PBR directory a given mseid is
    
    mse_num = int(mse[3:]) 
    if mse_num <= 4500:
        output_dir = '/data/henry7/PBR/subjects'
    else:
        output_dir = '/data/henry11/PBR/subjects'
    return output_dir
```

## block checks to see what exists from the df... luckily can always be run to check what's up

```python
"""lambda is like defining a function, but simpler and it only needs one line. You can look up syntax on line
syntax: 
lambda argument_list: expression
where argument list is separated by commas""" 

exist = lambda mseid, wf: os.path.exists(os.path.join(_get_output(mseid), mseid, wf))

wf_columns = ['mseID','nii', 'alignment', 'lst', 'mindcontrol', 'antsCT', 'siena', 'sienax']

# make new data frame with columns = wf_columns
new_df = pd.DataFrame(columns=wf_columns)


# make the 'mseID' column of new_df equal to the 'mseID' column of df
new_df['mseID'] = df['mseID']


for _,row in new_df.iterrows():
    # iterate thru each row of the new_df
    
    # store the mseid value
    mseid = new_df.loc[_, 'mseID']
    print(mseid)
    
    # now iteratre thru each column of the wf_columns list
    for workflow in wf_columns:
        
        # if column exists ( look at lambda function for definition of exists) then assign a value to new_df at [_,workflow]
        if exist(mseid, workflow):
            new_df.loc[_, workflow] = True 
```

```python
#print the length of new_df
print('Len of new_df {}'.format(len(new_df)))

# print number of antsCT that exist
print('Number of antsCT that exist {}'.format(len(new_df[new_df['antsCT'] == True])))

# print number of antsCT that don't exist
print('Number of antsCT that do not exist {}'.format(len(new_df[pd.isnull(new_df['antsCT'])])))
```
