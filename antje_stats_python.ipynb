Running a linear model

```python
formula = "Q('Left-Cerebellum-Cortex') ~ 1 + Q('Left-Thalamus-Proper')"  
```

```python
ols_object = sf.ols(formula,df)
```

```python
ols_object = sf.ols(formula,df)
```

```python
res = ols_object.fit()
```

```python
res.summary()
```

```python
res.tvalues #["Q('Left-Thalamus-Proper')"]
```

```python
t_stat = res.tvalues["Q('Left-Thalamus-Proper')"]
```
